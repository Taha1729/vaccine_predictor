```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
x_train = pd.read_csv('training_set_features.csv')
y_train= pd.read_csv('training_set_labels.csv')
x_test = pd.read_csv('test_set_features.csv')
y_test = pd.read_csv('submission_format.csv')
```

```python
x_train.shape
x_train.set_index('respondent_id', inplace = True)
x_test.set_index('respondent_id', inplace = True)
y_train.set_index('respondent_id', inplace = True)
y_test.set_index('respondent_id', inplace = True)
```

```python
x_train = x_train.fillna({'employment_industry':'none', 'employment_occupation':'unemployed', 'employment_status':'Not in Labor Force'})
pd.set_option('display.max_colwidth', 100)
```

```python
(x_train[x_train.isnull().any(axis=1)])
```

```python
feature_cols = [col for col in x_train.columns]
```

```python
imputer = SimpleImputer(strategy='most_frequent')
x_train[feature_cols] = imputer.fit_transform(x_train[feature_cols])
```

```python
x_test[x_test.isnull().any(axis=1)]
```

```python
x_test = x_test.fillna({'employment_industry':'none', 'employment_occupation':'unemployed', 'employment_status':'Not in Labor Force'})
```

```python
feature_cols2 = [col for col in x_test.columns]
x_test[feature_cols2] = imputer.fit_transform(x_test[feature_cols2])
```

```python
x_test['age_group'].unique()
```

```python
x_test.replace({'35 - 44 Years':40, '18 - 34 Years':26, '55 - 64 Years':60, '65+ Years':82,'45 - 54 Years':50}, inplace = True)
```

```python
x_train.replace({'35 - 44 Years':40, '18 - 34 Years':26, '55 - 64 Years':60, '65+ Years':82,'45 - 54 Years':50}, inplace = True)
```

```python
x_train['education'].unique()
```

```python
x_train.replace({'< 12 Years':0, '12 Years':1, 'College Graduate':3, 'Some College':2,'Male':0,'Female':1}, inplace = True)
```

```python
x_test.replace({'< 12 Years':0, '12 Years':1, 'College Graduate':3, 'Some College':2,'Male':0,'Female':1}, inplace = True)
```

```python
cat_col = ['race', 'sex', 'income_poverty', 'marital_status', 'rent_or_own', 'employment_status', 'hhs_geo_region', 'census_msa', 'employment_industry', 'employment_occupation']
```

```python
x_train[cat_col] = x_train[cat_col].astype('category')
x_train[cat_col] = x_train[cat_col].apply(lambda x: x.cat.codes)
x_test[cat_col] = x_test[cat_col].astype('category')
x_test[cat_col] = x_test[cat_col].apply(lambda x: x.cat.codes)
```

```python
from sklearn import linear_model
lasso = linear_model.Lasso(alpha = 50, max_iter = 100, tol = 0.1)
lasso.fit(x_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(max_iter = 1000)
model1.fit(x_train, y_train['xyz_vaccine'])
```

```python
pred1 = model1.predict_proba(x_test)[:,1]
y_test['h1n1_vaccine'] = pred1
```

```python
model2 = LogisticRegression(max_iter = 1000)
model2.fit(x_train, y_train['seasonal_vaccine'])
pred2 = model2.predict_proba(x_test)[:,1]
y_test['seasonal_vaccine'] = pred2
```
